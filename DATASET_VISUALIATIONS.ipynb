# Data Visualization

```python
import pandas as pd
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("Dataset.csv")
```

```python
data.head(15)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("DATASET OVERVIEW")
# Line chart showing daily global streams of each song 
sns.lineplot(data=data)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,6))

# Add title
plt.title("TIME VALUES")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=data.index, y=data["TIME"])

# Add label for vertical axis
plt.ylabel("TIME")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,6))

# Add title
plt.title("LOOPS AND THEIR CORRESPONDING UNROLLING FACTORS")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=data['LOOPS'], y=data["UF"])

# Add label for vertical axis
plt.ylabel("UNROLLING FACTORS")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,10))
sns.scatterplot(x=data[500:]['LOOPS'], y=data[500:]['TIME'])
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,6))

# Add title
plt.title("TIME AND UNROLLING FACTOR")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=data.loc[data.LOOPS == 28000]['UF'], label="UF")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data= data.loc[data.LOOPS == 28000]['TIME'], label="TIME")

# Add label for horizontal axis
plt.xlabel("DS ITEMS")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,5))
sns.scatterplot(x=data[data.LOOPS == 20000]['TIME'], y=data[data.LOOPS == 20000]['UF'])
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,5))
sns.scatterplot(x=data[data.LOOPS == 20000]['TIME'], y=data[data.LOOPS == 20000]['UF'])
```
